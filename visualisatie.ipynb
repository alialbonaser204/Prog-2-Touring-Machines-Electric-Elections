```python
import pandas as pd

# Lees het bestand
df = pd.read_csv(r"data\per_uur_data_multi_run1.csv")
df.head()
```

```python
plt.figure(figsize=(12, 6))

# Wachttijden per richting per uur plotten
plt.plot(df["uur"], df["wachttijd_rechts"], marker='o', label='Rechts')
plt.plot(df["uur"], df["wachttijd_links"], marker='s', label='Links')
plt.plot(df["uur"], df["wachttijd_rechtdoor"], marker='^', label='Rechtdoor')

# Labels en titel
plt.title('Gemiddelde wachttijd per richting per uur (Multi Run 1)')
plt.xlabel('Uur van de dag')
plt.ylabel('Gemiddelde wachttijd (minuten)')
plt.legend()
plt.grid(True)
plt.xticks(df["uur"]) 
plt.tight_layout()

plt.show()
```
